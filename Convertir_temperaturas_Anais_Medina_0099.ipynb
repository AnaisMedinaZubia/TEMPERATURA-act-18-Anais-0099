## 1. Fórmulas de Conversión

Para convertir entre estas escalas, utilizamos relaciones lineales basadas en los puntos de congelación ($0$ °C / $32$ °F) y ebullición ($100$ °C / $212$ °F) del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(O su equivalente decimal: $°F = (°C \times 1.8) + 32$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$
    *(O su equivalente fraccionario: $°C = (°F - 32) \times \frac{5}{9}$)*

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25 °C a Fahrenheit
**Objetivo:** Saber a cuánto equivalen 25 grados centígrados (una temperatura ambiente agradable).

1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$.
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar la constante 32:** $45 + 32 = 77$.
* **Resultado:** $25$ °C equivalen a **$77$ °F**.

### Ejemplo B: Convertir 98.6 °F a Celsius
**Objetivo:** Convertir la temperatura corporal promedio humana.

1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$.
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$.
3.  **Restar la constante 32:** $98.6 - 32 = 66.6$.
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$.
* **Resultado:** $98.6$ °F equivalen a **$37$ °C**.

## 3. Código en Python

Aquí tienes el código estructurado de forma modular, utilizando funciones y una estructura de control interactiva.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    """Muestra el menú principal y gestiona la lógica del programa"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES DE TEMPERATURA ---")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Por favor, intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`float()`**: Se usa para permitir que el usuario ingrese números con decimales (ej. 36.5).
* **`while True`**: Crea un bucle infinito que solo se rompe cuando el usuario elige la opción 3 (**`break`**).
* **`f-strings`**: Se utilizan en los `print` para insertar las variables y limitar los decimales a dos con `:.2f`.
* **Modularidad**: Al separar las conversiones en funciones (`convertirCF` y `convertirFC`), el código es más limpio y fácil de mantener.
